Create a 2D numpy array A and a 1D numpy array
$$ A = \begin{bmatrix}
1 & 2 & 3 \\
4 & 5 & 6
\end{bmatrix}\,, 
x = \begin{bmatrix}
1 \\ -1 \\ 2
\end{bmatrix}
$$
and compute $Ax$.

```python
#CODE HERE
A = np.array([[1,2,3], [4,5,6]])
x = np.array([1,-1,2])
A*x
```

Construct a function that multipies a matrix times a vector

```python
def matrix_vector_mult(A,x):
    """
    This function returns the product of the matrix A and vector x.
    The product is computed by looping over columns only.
    There is no check that the the number of columns of A is equal to the number of elements of x
    """
    nrows, ncolumns = A.shape
    #CODE HERE
    b = np.array([0,0])
    for ii in range(0,ncolumns):
        b = b + A[:,ii]*x[ii]
    return b
matrix_vector_mult(A,x)
```

Compute the climatology of Nino 3.4 in 3 different ways and get 3 different answers.

```python
# 41 years
nmonths = n34.size
nmonths/12
```

Method 1 (loops)

```python
climatology = np.full(12,np.nan)
anomaly = np.zeros_like(n34)
```

```python
for imonth in range(1,12):
    months = range(imonth, nmonths, 12)
    climatology[imonth] = n34[months].mean()
    anomaly[months] = n34[months] - n34[months].mean() #CODE
    
#problem here was that parameters of range function were reversed. you said start at imonth, stop at 12, and iterate by 492.
#Need to start at imonth, stop at 492, and iterate by 12.
```

Method 2 (reshape)

```python
# reshape the data so that each of the 41 columns contains a year and then average over columns (years) 
climatology2 = n34.values.reshape(41,12).mean(axis = 0)
anomaly2 = n34.values.reshape(41,12) - climatology2
anomaly2.resize(nmonths)
climatology2

#We were trying to average over 41 columns, each containing a year, but reshaping fills rows in order.
#So the first row would contain the first 41 months, rather than all 41 januarys.
```

Method 3 (xarray groupby)  
http://xarray.pydata.org/en/stable/groupby.html

```python
climatology3 = n34.groupby("T.month").mean()
anomaly3 = n34.groupby("T.month") - climatology3
```
